# A simple 1D differential equation
As a first example, I try to solve the diffusion equation, i.e., $$ \frac {\partial c} {\partial t} + \nabla.(-D \nabla c)=0$$ The boundary condition is written in the general form of $$a \nabla \phi . \mathbf{n} + b \phi = c$$

```julia
# The Julia implementation
L = 1.0 # Domain length
Nx = 10 # number of cells [uniform]
m = createMesh1D(Nx, L) # 1D Cartesian mesh
m = createMesh2D(2*Nx, Nx, 2*L, L) # 2D Cartesian mesh
m = createMeshRadial2D(3*Nx, 2*Nx, L, 2*pi) # 2D Radial mesh
m.cellcenters.x+=0.1
m.facecenters.x+=0.1
BC = createBC(m) # creates the boundary condition structure [default Neumann]
# change left boundary to Dirichlet
BC.left.a[:] = 0
BC.left.b[:] = 1.0
BC.left.c[:] = 1.0
# change right boundary to Dirichlet
BC.right.a[:] = 0
BC.right.b[:] = 1.0
BC.right.c[:] = 0.0
# Define initial condition
c0 = 0.0
c_old = createCellVariable(m, c0, BC)
# Define the transfer coeffiicient
D_val = 0.01 # diffusion coefficient value (can be different on each cell)
D = createCellVariable(m, D_val)
D_face = arithmeticMean(D) # average on the cell faces
# Construct the discretized PDE
Mdiff = diffusionTerm(D_face)
Mbc, RHSbc = boundaryConditionTerm(BC)
# time step
dt = 1.0
final_t = 10.0
# solver loop
for t=dt:dt:final_t
    Mt, RHSt = transientTerm(c_old, dt)
    c_new = solveLinearPDE(m, Mbc-Mdiff+Mt, RHSt+RHSbc)
    c_old = copyCell(c_new)
end
visualizeCells(c_old)
#println(c_new)
```
